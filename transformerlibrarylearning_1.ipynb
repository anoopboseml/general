```python
# we will import pipeline functionality from Hugging Face's Transformer library
from transformers import pipeline

# we will then use the pipeline module to tell what task we want to perform
classifier = pipeline(task = 'sentiment-analysis')
```

```python
text = "I loved this movie"

result = classifier(text)
print(result)
```

```python
# pipeline auto used the model distilbert/distilbert-base-uncased-finetuned-sst-2-english
# you can not/should not do this in a production environment
```

```python
text = "This movie was horseshit"

result = classifier(text)
print(result)
```

```python
text = "it was an average movie"

result = classifier(text)
print(result)
```

```python
texts = ["I loved this movie", 'It was a really bad film', 'It should have been rated 0 instead of 10']

results = classifier(text)
print(results)
```

```python
# we can see the above result in a neater way
for review, result in zip(texts, results):
  print(f'Review: {review}')
  print(f'Sentiment: {result["label"]}')
  print(f'Score: {result['score']}')
```
